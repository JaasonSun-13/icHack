```python
import pandas as pd
import numpy as np
import yfinance as yf
import time
from pytrends.request import TrendReq
```

```python

pytrends = TrendReq(hl='en-US', tz=360)


pytrends.build_payload(
        ['wordle'],
        timeframe=f'{2022}-01-01 {2022}-12-31'
    )

df = pytrends.interest_over_time()

df.to_csv(f'FinanceDataPuller/tester/tester.csv')
```
